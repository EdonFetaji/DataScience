```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/EdonFetaji/DataScience/refs/heads/main/dataSets/email.csv')
```

```python
data.head()
```

```python
encoder = LabelEncoder()
data['Category'] = encoder.fit_transform(data['Category'])
```

```python
data.sample(5)
```

```python
vectorizer = CountVectorizer(stop_words='english')
```

```python
x = data['Message']
x_cv = vectorizer.fit_transform(x)
y = data['Category']
```

```python
x_cv.toarray()
```

```python
# train_x_cv = vectorizer.fit_transform(train_x)
# test_x_cv = vectorizer.transform(test_x)
```

```python
train_x,test_x,train_y,test_y = train_test_split(x_cv.toarray(),y,test_size= 0.2,random_state=42)
```

```python
def execute_model(model):
  model.fit(train_x,train_y)
  return  model.predict(test_x)
```

```python
def evaluate_prediction(pred_y):
  print(classification_report(test_y,pred_y))
```

```python
predict_m1_y= execute_model(MultinomialNB())
```

```python
evaluate_prediction(predict_m1_y)
```

```python
evaluate_prediction(execute_model(DecisionTreeClassifier()))
```

```python
evaluate_prediction(execute_model(LogisticRegression()))
```

```python
evaluate_prediction(execute_model(LogisticRegression()))
```

# TF-IDF(t,d) = TF(t,d) √ó IDF(t,D)

```python
import spacy
```

```python
nlp  = spacy.load("en_core_web_sm")
```

```python
def process_text(text):
  doc = nlp(text)
  filtered = []
  for token in doc:
    if token.is_punct:
      continue
    filtered.append(token.lemma_)
  return " ".join(filtered)
```

```python
tester = "Hello this is just a test word, to see if my method for lemmatization correctly works"

print(process_text(tester))
```

```python
data['processed_msg']= data.Message.apply(process_text)
```

```python
train_x,test_x,train_y,test_y = train_test_split(data.processed_msg,data.Category,test_size=0.2,random_state=69)
```

```python
train_x_v = vectorizer.fit_transform(train_x).toarray()
test_x_v = vectorizer.transform(test_x).toarray()
```

```python
model = MultinomialNB()
model.fit(train_x_v ,train_y)
pred_y = model.predict(test_x_v)
```

```python
evaluate_prediction(test_y,pred_y)
```

```python
fake = pd.read_csv('Fake.csv')
fake = fake[['text']]
fake['Category']  = pd.Series([0] * len(fake))
fake.sample(10)
```

```python
true = pd.read_csv('True.csv')
true = true[['text']]
true['Category']  = pd.Series([1] * len(true))
true.sample(10)
```

```python
df = pd.concat([true,fake])
df.sample(15)
```

```python
import spacy
!python -m spacy download en_core_web_lg
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
df['text_vectorized'] = df['text'].apply(lambda x: nlp(x).vector)
```

```python
train_x,test_x,train_y,test_y = train_test_split(df.text_vectorized.values,df.Category,test_size= 0.2, shuffle=True , random_state=69)
```

```python
import numpy as np
```

```python
train_x = np.stack(train_x)
test_x = np.stack(test_x)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
train_x_s = scaler.fit_transform(train_x)
test_x_s = scaler.fit_transform(test_x)
```

```python
model =  MultinomialNB()
model.fit(train_x_s,train_y)
```

```python
pred_m_y = model.predict(test_x_s)
```

```python
print(classification_report(test_y,pred_m_y))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5,metric='euclidean')
```

```python
knn.fit(train_x,train_y)
```

```python
print(classification_report(test_y,knn.predict(test_x)))
```
